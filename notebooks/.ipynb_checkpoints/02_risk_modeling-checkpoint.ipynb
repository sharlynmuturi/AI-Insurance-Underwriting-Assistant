The goal is to build an underwriting model that:
- Predicts claim probability (frequency)
- Produces a calibrated risk score
- Supports pricing decisions

```python
import pandas as pd

df = pd.read_csv("insurance_dataset.csv")
df.head()
```

```python
df.columns
```

### Feature Selection
Removing leakage columns
- premiums depends on risk_score
- claim_amount, claim_type and risk_score only exists after claim

```python
drop_cols = ["claim_occurred", "claim_amount", "premium", "base_premium", "risk_score", "claim_type"]

X = df.drop(columns=drop_cols)
y = df["claim_occurred"]
```

### Feature Engineering

We create underwriting factors by converting raw features

```python
# Safety Score - Measures how well-protected a vehicle is against accidents and theft
# Giving tracking device more impact
X["safety_score"] = (X["airbags"] * 0.5 + X["tracking_device"] * 2)
```

```python
# Behavioral Risk Score - Captures how risky a driver’s habits are based on usage and history
speed_map = {"low":0, "medium":1, "high":2}

X["behavior_score"] = (X["annual_mileage"] / 10000 + X["previous_claims"] * 2 + X["speeding_score"].map(speed_map))
```

```python
# Age Buckets - Categorical grouping to capture non-linear effects of age on driving risk.
X["age_band"] = pd.cut(X["age"], bins=[20,30,50,70], labels=["young","mid","older"])
```

```python
# Vehicle Age Bands - to capture non-linear effects of vehicle condition on claim likelihood.
X["vehicle_age_band"] = pd.cut(X["vehicle_age"], bins=[0,3,7,15], labels=["new","mid","old"])
```

```python
# Mileage Risk - Binary indicator to highlight drivers with high road exposure and higher accident probability.
X["high_mileage"] = (X["annual_mileage"] > 25000).astype(int)
```

```python
X.columns
```

```python
# Encoding Categoricals
X = pd.get_dummies(X, drop_first=True)
```

```python
# Checking correlation to ensure no more leakage
import numpy as np

for col in X.columns:
    corr = np.corrcoef(X[col], y)[0,1]
    print(col, corr)
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# Class Imbalance

print(y_train.value_counts(normalize=True))
```

### Model Training
#### Logistic Regression

```python
X.describe()
```

There are huge differences in scale:

- vehicle_value - up to ~7,000,000
- annual_mileage - up to 40,000
- age - ~20–70
- tracking_device - 0–1

Scaling the values so lr is not biased towards large-value features

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

lr = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

lr.fit(X_train, y_train)
```

```python
coefficients = pd.Series(lr.named_steps["model"].coef_[0], index=X.columns)
```

```python
top_risk = coefficients.sort_values(ascending=False).head(3)
top_safe = coefficients.sort_values().head(3)

print("Top risk drivers:\n", top_risk)
print("Risk reducers:\n", top_safe)
```

```python
y_pred_proba = lr.predict_proba(X_test)[:,1]

y_pred = (y_pred_proba > 0.6).astype(int)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba)
```

```python
# Plotting True Positive Rate (Recall) vs False Positive Rate at every possible threshold.

import matplotlib.pyplot as plt

from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_pred_proba)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.title("ROC Curve")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

- 8501 (TN) - correctly predicted no claim
- 661 (TP) - predicted claim correctly
- 743 (FN) - missed claims
- 95 (FP) - predicted claim incorrectly

```python
from sklearn.metrics import classification_report

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

##### LR Risk Score Creation

```python
df_test = X_test.copy()
df_test["actual"] = y_test.values

df_test["lr_predicted_risk"] = y_pred_proba

df_test["lr_risk_band"] = pd.cut(df_test["lr_predicted_risk"], bins=[0,0.05,0.1,0.2,1], labels=["Low","Medium","High","Very High"])

df_test.groupby("lr_risk_band", observed=True)["actual"].mean()
```

0nly ~8% of policies in “Very High” actually claimed. This is much lower than what we’d expect for a high-risk band.
- The logistic regression isn’t separating the extremes well.

Trying a Gradient Boosting model that handles imbalance automatically

### Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingClassifier
gbm = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42)

gbm.fit(X_train, y_train)
```

```python
importances = pd.Series(gbm.feature_importances_, index=X.columns)
importances.sort_values(ascending=True).head(15).plot(kind="barh", figsize=(10,6))
plt.title("Top 15 Feature Importances - Gradient Boosting")
plt.show()
```

```python
gbm_pred_proba = gbm.predict_proba(X_test)[:,1]
y_pred = (gbm_pred_proba > 0.4).astype(int)
```

```python
confusion_matrix(y_test, y_pred)
```

- 9145 (TN) - correctly predicted no claim
- 17 (TP) - predicted claim correctly
- 836 (FN) - missed claims
- 2 (FP) - predicted claim incorrectly

```python
from sklearn.metrics import roc_auc_score, classification_report

roc_auc_score(y_test, y_pred)
print("ROC-AUC:", roc_auc_score(y_test, gbm_pred_proba))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.title("ROC Curve")
plt.show()
```

- The model is not separating fraud vs non-fraud well at all (recall - 0.00)

```python
df_test["gb_predicted_risk"] = gbm_pred_proba

df_test["gb_risk_band"] = pd.cut(df_test["gb_predicted_risk"], bins=[0,0.05,0.1,0.2,1], labels=["Low","Medium","High","Very High"])

df_test.groupby("gb_risk_band", observed=True)["actual"].mean()
```

- The model performs worse than Logistic Regression model (recall - 0.00).

Training a Random Forest model with SMOTE for imbalance handling

#### Random Forest

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, class_weight="balanced", max_depth=10, random_state=42)
rf.fit(X_train_res, y_train_res)
```

```python
importances = pd.Series(rf.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).head(10)
```

```python
rf_pred = rf.predict_proba(X_test)[:,1]

y_pred = (rf_pred > 0.4).astype(int)
```

```python
roc_auc_score(y_test, rf_pred)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.title("ROC Curve")
plt.show()
```

```python
confusion_matrix(y_test, y_pred)
```

- 8833 (TN) - correctly predicted no claim
- 329 (TP) - predicted claim correctly
- 789 (FN) - missed claims
- 49 (FP) - predicted claim incorrectly

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

##### RF Risk Score Creation

```python
df_test["rf_predicted_risk"] = rf_pred

df_test["rf_risk_band"] = pd.qcut(df_test["rf_predicted_risk"], 4, labels=["Low","Medium","High","Very High"])

df_test.groupby("rf_risk_band", observed=True)["actual"].mean()
```

- shows a more logical monotonic trend(“higher risk band - higher fraud rate”)
- Better than GBM, still lower recall than the LR Model.

Training an XGBoost model

- Handles non-linear relationships better than Logistic Regression
- Better than Random Forest for structured/tabular data
- Has built-in handling for class imbalance (scale_pos_weight)

#### XGBoost Model

```python
# Computing imbalance ratio

scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(scale_pos_weight)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    eval_metric="logloss"
)

xgb_model.fit(X_train, y_train)
```

```python
# Finding optimal threshold

from sklearn.metrics import f1_score

thresholds = np.arange(0.1, 0.9, 0.05)

for t in thresholds:
    preds = (y_pred_proba > t).astype(int)
    print(t, f1_score(y_test, preds))
```

- Either 0.4 or 0.5 (0.4 produces the better recall)

```python
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]
y_pred = (y_pred_proba > 0.4).astype(int)

print("ROC-AUC:", roc_auc_score(y_test, y_pred_proba))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.title("ROC Curve")
plt.show()
```

```python
df_test["xgb_risk"] = y_pred_proba

df_test["risk_band"] = pd.qcut(df_test["xgb_risk"], q=4, labels=["Low", "Medium", "High", "Very High"])

print(df_test.groupby("risk_band", observed=True)["actual"].mean())
```

- Clean monotonic increase - Very High has highest fraud rate (useful for pricing + underwriting)

```python
importance = pd.Series(xgb_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)

print(importance.head(10))
```

- region_Nairobi - urban risk
- speeding_score_medium - behavior
- vehicle_type_truck/pickup - exposure
- policy_duration - short-term risk
- behavior_score, safety_score

This is aligned with real-world insurance logic

##### Conclusion

XGBoost Model outperforms the other models with a recall of 0.68 (model catches ~68% of fraud).

- Risk bands are meaningful
- Feature importance makes sense
- Probabilities are more spread out

### Pricing Model (Expected Loss and Premium)

- Expected loss (insurer’s anticipated claim cost) = model’s estimated risk * vehicle value.
- Technical premium - ensures operational costs are covered and profitability achieved
  
      - Expense loading - operational costs added on top of expected claims (comes from historical company data)
      - profit_margin - insurer’s target return on risk. Accounting for shareholder return, solvency requirements and risk buffer
  
Setting Expense and profit loadings at 30% and 10% respectively, to reflect typical motor insurance market ranges

```python
df_test["expected_loss"] = df_test["xgb_risk"] * df.loc[df_test.index, "vehicle_value"]

expense_loading = 0.3
profit_margin = 0.1

df_test["technical_premium"] = df_test["expected_loss"] * (1 + expense_loading + profit_margin)
```
